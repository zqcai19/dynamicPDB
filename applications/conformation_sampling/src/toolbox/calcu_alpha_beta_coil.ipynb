```python
def compute_secondary_structure_proportions(pdb_path):
    """
    Compute the proportions of alpha helices, beta sheets, and coils in the given PDB file.

    Args:
    pdb_path (str): Path to the PDB file.

    Returns:
    tuple: Proportions of alpha helices, beta sheets, and coils.
    """
    traj = md.load(pdb_path)
    dssp = md.compute_dssp(traj)
    total_residues = dssp.shape[1]

    # Flatten the DSSP assignments and count occurrences
    dssp_flat = dssp.flatten()
    alpha_helix_count = np.sum(dssp_flat == 'H')
    beta_sheet_count = np.sum(dssp_flat == 'E')
    coil_count = np.sum((dssp_flat == 'C') | (dssp_flat == 'S') | (dssp_flat == 'T') | (dssp_flat == 'G') | (dssp_flat == 'B') | (dssp_flat == 'I'))

    # Calculate proportions
    alpha_helix_proportion = alpha_helix_count / len(dssp_flat)
    beta_sheet_proportion = beta_sheet_count / len(dssp_flat)
    coil_proportion = coil_count / len(dssp_flat)

    return alpha_helix_proportion, beta_sheet_proportion, coil_proportion
```

# compute_secondary_structure_proportions

```python
atlas_data[['alpha_helix_proportion', 'beta_sheet_proportion', 'coil_proportion']] = atlas_data['pdb_path'].apply(
    lambda x: pd.Series(compute_secondary_structure_proportions(x))
)
atlas_data.head(2)
```
